```
x=np.linspace(0,10,20)
```

```
y=np.sort(y)
```

```
plt.plot(x,y,color='m',marker='*',linestyle='--',markersize=10,lw=1.5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('x vs y')
plt.show()
```

```
plt.subplot(2,2,1)
plt.plot(x,y, color='m',marker='*',markersize=2)
plt.subplot(2,2,2)
plt.plot(x,y, color='b')
plt.subplot(2,2,3)
plt.plot(x,y, color='g')
plt.subplot(2,2,4)
plt.plot(x,y, color='r')
plt.show()
```

##                     ________ object oriented                





```
import numpy as np 
import matplotlib.pyplot as plt
from numpy.random import randint
```

```
x=np.linspace(0,10,20)
y=randint(0,50,20)
y=np.sort(y)
```

```
fig= plt.figure()
#4 coordinates are left,bottom,height,width
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(x,y,color='m')
```

```
# multiple subplot
fig= plt.figure()
#4 coordinates are left,bottom,height,width and are in percentage value lie betn 0 -1
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])
ax1.plot(x,y,color='m')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.plot(x,y,color='b')
fig.show()
```

# Figure size aspect ratio

```
import numpy as np 
import matplotlib.pyplot as plt
from numpy.random import randint
```

```
x=np.linspace(0,10,20)
y=randint(0,50,20)
y=np.sort(y)
```

```
x
```

```
y
```

```
fig=plt.figure(figsize=(4,3))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```
# multiple subplots
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(6,3))
fig.tight_layout()
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```
fig
```

```
fig.savefig('myfirst.png',dpi=100)
```

```
# specify legends
```

```
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x vs x^2")
ax.plot(x,x**3,label="x vs x^3")
ax.legend()
```

#  3) different plots 

##      1) scatter plot

```
# scatter plot
import matplotlib.pyplot as plt
```

```
f_view=[100,250,400,450,500,300,250]
y_view=[50,150,190,390,460,500,250]
t_view=[75,130,400,450,650,450,600]
days=[1,2,3,4,5,6,7]
```

```
plt.scatter(days,f_view,label='facebook view',marker='o')
plt.scatter(days,y_view,label='youtube view',marker='o')
plt.scatter(days,t_view,label='youtube view',marker='o')
plt.legend()
plt.show()
```

##       2) Bar plot

```
plt.bar(days,f_view)
plt.bar(days,t_view)
plt.show()
```
